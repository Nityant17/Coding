```python
'''
For the MATCHES dataset, answer the specified questions with summarization and effective visuals.
1. Find the names of teams that are part of IPL. 
2. Visualize the total number of matches won by each team. 
3. List the names of umpires who have served as third umpires and the number of matches. 
4. Visualize how many matches were held in each venue. 
5. Visualize the top 10 players identified as the player of the match for the most significant number of matches. 
6. Visualize the bottom 10 venues where the least number of matches were played. 
7. Visualize the number of wins of each team for the venue of ‘MA Chidambaram Stadium, Chepauk’ 
8. Find and list the venues where the match resulted in a tie. 
9. List and visualize the best teams(top 5), season-wise.  
10. Analyse and visualize the relationship between winning the toss and winning the match.
'''
 # Week 7, Code 2
 # Nityant Agarwal, CC, 240953418
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('matches.csv')
df
```

```python
# 1. Find the names of teams that are part of IPL
teams = df['team1'].unique()
print("Teams in IPL:", teams)
```

```python
# 2. Visualize the total number of matches won by each team
wins = df['winner'].value_counts()
wins.plot(kind='bar', figsize=(10,5), color='skyblue')
plt.title("Total Matches Won by Each Team")
plt.xlabel("Teams")
plt.ylabel("Matches Won")
plt.xticks(rotation=70)
plt.show()
```

```python
# 3. List the names of umpires who have served as third umpires and the number of matches
umpires = df['umpire3'].value_counts()
print(umpires)
print("Third Umpires and Matches Served: None")
```

```python
# 4. Visualize how many matches were held in each venue
venue_counts = df['venue'].value_counts()
venue_counts.plot(kind='bar', figsize=(12,6), color='orange')
plt.title("Matches Held at Each Venue")
plt.xlabel("Venue")
plt.ylabel("Number of Matches")
plt.xticks(rotation=90)
plt.show()
```

```python
# 5. Visualize the top 10 players identified as the player of the match
pom_counts = df['player_of_match'].value_counts().head(10)
pom_counts.plot(kind='bar', figsize=(10,5), color='green')
plt.title("Top 10 Players - Player of the Match Awards")
plt.xlabel("Players")
plt.ylabel("Awards Won")
plt.xticks(rotation=45)
plt.show()
```

```python
# 6. Visualize the bottom 10 venues where the least number of matches were played
venue_counts.tail(10).plot(kind='bar', figsize=(10,5), color='red')
plt.title("Bottom 10 Venues - Least Number of Matches Played")
plt.xlabel("Venue")
plt.ylabel("Number of Matches")
plt.xticks(rotation=90)
plt.show()
```

```python
# 7. Visualize the number of wins of each team for MA Chidambaram Stadium
chepauk_wins = df[df['venue'] == 'MA Chidambaram Stadium, Chepauk']['winner'].value_counts()
chepauk_wins.plot(kind='bar', figsize=(10,5), color='purple')
plt.title("Wins at MA Chidambaram Stadium, Chepauk")
plt.xlabel("Teams")
plt.ylabel("Wins")
plt.xticks(rotation=70)
plt.show()
```

```python
# 8. Find and list venues where the match resulted in a tie
tie_venues = df[df['result'] == 'tie']['venue'].unique()
print("Venues where matches resulted in a tie:\n", tie_venues)
```

```python
# 9. List and visualize the best teams (top 5), season-wise
season_winners = df.groupby(['season', 'winner']).size().reset_index(name='wins')
top_5_season_winners = season_winners.sort_values(by=['season', 'wins'], ascending=[True, False]).groupby('season').head(5)
plt.figure(figsize=(15,15))
sns.barplot(data=top_5_season_winners, x='season', y='wins', hue='winner', palette='muted')
plt.title("Top 5 Teams Season-wise")
plt.xticks(rotation=45)
plt.show()
```

```python
# 10. Analyse and visualize the relationship between winning the toss and winning the match
df['toss_win_match_win'] = df['toss_winner'] == df['winner']
toss_impact = df['toss_win_match_win'].value_counts(normalize=True) * 100
toss_impact.plot(kind='bar', color=['blue', 'gray'], figsize=(8,5))
plt.title("Impact of Winning Toss on Winning the Match")
plt.xticks([0,1], ['Lost Match', 'Won Match'], rotation=0)
plt.ylabel("Percentage")
plt.show()
```
