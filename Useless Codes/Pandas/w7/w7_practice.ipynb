```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('mtcars.csv')
df
```

```python
# 1. Extract the 'mpg' column from the dataframe & Plot the mpg values as a line plot
mpg = df['mpg']
mpg.plot()
plt.xlabel("Index")
plt.ylabel("MPG")
plt.title("MPG values")
plt.show()
```

```python
# 2. Select only the 'cyl', 'wt', and 'mpg' columns from the dataframe & Plot the selected dataframe columns
df = df[['cyl', 'wt', 'mpg']]
df.plot()
plt.xlabel("Index")
plt.ylabel("Values")
plt.show()
```

```python
#3. Bar Plot- the count of cars by the number of cylinders
cyl_counts=df["cyl"].value_counts()
plt.figure(figsize=(5,5))
sns.barplot(x=cyl_counts.index, y=cyl_counts.values,palette="viridis")
plt.title('Count of Cars by Cylinders')
plt.xlabel('Cylinders')
plt.ylabel('Count')
plt.show()
```

```python
#use ggplot for better visualization
plt.style.use('ggplot')
```

```python
# 4. Give scatter plot of weight vs mpg
plt.figure(figsize=(9, 9))
sns.scatterplot(x=df["wt"], y=df["mpg"], alpha=0.7)
plt.xlabel("Weight")
plt.ylabel("MPG")
plt.title("Relationship Between Weight and MPG")
plt.grid(True)
plt.show()
```

```python
# 5. Box plot- MPG for each cylinder category
plt.figure(figsize=(8, 6))
sns.boxplot(x='cyl', y='mpg', data=df)
plt.title('MPG by Cylinder Category')
plt.xlabel('Number of Cylinders')
plt.ylabel('MPG')
plt.show()
```

```python
# 6. Give Histogram of MPG for all cars
x = df['mpg']
plt.hist(x)
plt.title("Distribution of MPG in Cars Dataset")
plt.ylabel('Frequency')
plt.xlabel('MPG')
plt.show()
```

```python
# 7. Give pie chart that shows the proportion of cars with 4, 6, or 8 cylinders
cyl_counts = df['cyl'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(
    cyl_counts,
    labels=cyl_counts.index,
    autopct='%1.1f%%',  
    colors=['skyblue', 'lightgreen', 'orange'],
    startangle=140,  
    explode=[0.1, 0, 0]  
)
plt.title('Proportion of Cars by Number of Cylinders')
plt.show()
```

```python
# 8. Give Pie Chart for Transmission Types
am_counts = df["am"].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(
    am_counts,
    labels=['Automatic', 'Manual'],  
    autopct='%1.1f%%',
    colors=['gold', 'lightcoral'],
    startangle=90
)
plt.title('Proportion of Cars by Transmission Type')
plt.show()
```

```python
# 9. Give a Nested Pie Chart
outer_labels = df['cyl'].value_counts().index
outer_sizes = df['cyl'].value_counts().values
inner_labels = df['gear'].value_counts().index
inner_sizes = df['gear'].value_counts().values
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(outer_sizes, labels=outer_labels, radius=1, autopct='%1.1f%%', colors=['skyblue', 'lightgreen', 'orange'])
ax.pie(inner_sizes, labels=inner_labels, radius=0.7, autopct='%1.1f%%', colors=['pink', 'purple', 'cyan'])
ax.set(aspect="equal")
plt.title('Nested Pie Chart: Cylinders (Outer) and Gears (Inner)')
plt.show()
```
